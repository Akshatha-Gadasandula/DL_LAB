```python
# Implement RNN for predicting the next character, word and sentence

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

text = "deep learning is powerful and deep learning is useful"

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

total_words = len(tokenizer.word_index) + 1

input_sequences = []
for i in range(1, len(text.split())):
    seq = tokenizer.texts_to_sequences([text.split()[:i+1]])[0]
    input_sequences.append(seq)

max_len = max(len(x) for x in input_sequences)

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_len, padding='pre')
)

X = input_sequences[:, :-1]
y = tf.keras.utils.to_categorical(
    input_sequences[:, -1],
    num_classes=total_words
)

model = Sequential([
    Embedding(total_words, 10, input_length=max_len - 1),
    SimpleRNN(64),
    Dense(total_words, activation='softmax')
])

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model.fit(X, y, epochs=100, verbose=0)

print("RNN Model Completed")
```

```python
# Implement LSTM and GRU architectures to deal with long term dependencies

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, GRU, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

text = "deep learning is powerful and deep learning is useful"

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

total_words = len(tokenizer.word_index) + 1

input_sequences = []
for i in range(1, len(text.split())):
    seq = tokenizer.texts_to_sequences([text.split()[:i+1]])[0]
    input_sequences.append(seq)

max_len = max(len(x) for x in input_sequences)

input_sequences = np.array(
    pad_sequences(input_sequences, maxlen=max_len, padding='pre')
)

X = input_sequences[:, :-1]
y = tf.keras.utils.to_categorical(
    input_sequences[:, -1],
    num_classes=total_words
)

lstm_model = Sequential([
    Embedding(total_words, 10, input_length=max_len - 1),
    LSTM(64),
    Dense(total_words, activation='softmax')
])

lstm_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

lstm_model.fit(X, y, epochs=100, verbose=0)

gru_model = Sequential([
    Embedding(total_words, 10, input_length=max_len - 1),
    GRU(64),
    Dense(total_words, activation='softmax')
])

gru_model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

gru_model.fit(X, y, epochs=100, verbose=0)

print("LSTM and GRU Models Completed")
```

```python
# Implement Encoder-Decoder Model for translation

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense

vocab_size = 1000
max_len = 10

encoder_inputs = Input(shape=(max_len,))
encoder_embedding = Embedding(vocab_size, 64)(encoder_inputs)

encoder = LSTM(64, return_state=True)
encoder_outputs, state_h, state_c = encoder(encoder_embedding)

encoder_states = [state_h, state_c]

decoder_inputs = Input(shape=(max_len,))
decoder_embedding = Embedding(vocab_size, 64)(decoder_inputs)

decoder_lstm = LSTM(
    64,
    return_sequences=True,
    return_state=True
)

decoder_outputs, _, _ = decoder_lstm(
    decoder_embedding,
    initial_state=encoder_states
)

decoder_dense = Dense(
    vocab_size,
    activation='softmax'
)

decoder_outputs = decoder_dense(decoder_outputs)

model = Model(
    [encoder_inputs, decoder_inputs],
    decoder_outputs
)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy'
)

print("Encoder-Decoder Model Completed")
```

```python
# Implement Attention Mechanism

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention, Concatenate

vocab_size = 1000
max_len = 10

encoder_inputs = Input(shape=(max_len,))
encoder_embedding = Embedding(vocab_size, 64)(encoder_inputs)

encoder_outputs, state_h, state_c = LSTM(
    64,
    return_sequences=True,
    return_state=True
)(encoder_embedding)

decoder_inputs = Input(shape=(max_len,))
decoder_embedding = Embedding(vocab_size, 64)(decoder_inputs)

decoder_outputs, _, _ = LSTM(
    64,
    return_sequences=True,
    return_state=True
)(
    decoder_embedding,
    initial_state=[state_h, state_c]
)

attention_output = Attention()(
    [decoder_outputs, encoder_outputs]
)

concat = Concatenate()(
    [decoder_outputs, attention_output]
)

output = Dense(
    vocab_size,
    activation='softmax'
)(concat)

model = Model(
    [encoder_inputs, decoder_inputs],
    output
)

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy'
)

print("Attention Model Completed")
```
